```python
import random

class BlackJack_game:
    def __init__(self):
        
        self.deck={'Ace of Hearts': 1, '2 of Hearts': 2, '3 of Hearts': 3, '4 of Hearts': 4, '5 of Hearts': 5,
            '6 of Hearts': 6, '7 of Hearts': 7, '8 of Hearts': 8, '9 of Hearts': 9, '10 of Hearts': 10,
            'Jack of Hearts': 10, 'Queen of Hearts': 10, 'King of Hearts': 10,
            
            'Ace of Diamonds': 1, '2 of Diamonds': 2, '3 of Diamonds': 3, '4 of Diamonds': 4, '5 of Diamonds': 5,
            '6 of Diamonds': 6, '7 of Diamonds': 7, '8 of Diamonds': 8, '9 of Diamonds': 9, '10 of Diamonds': 10,
            'Jack of Diamonds': 10, 'Queen of Diamonds': 10, 'King of Diamonds': 10,
            
            'Ace of Clubs': 1, '2 of Clubs': 2, '3 of Clubs': 3, '4 of Clubs': 4, '5 of Clubs': 5,
            '6 of Clubs': 6, '7 of Clubs': 7, '8 of Clubs': 8, '9 of Clubs': 9, '10 of Clubs': 10,
            'Jack of Clubs': 10, 'Queen of Clubs': 10, 'King of Clubs': 10,
            
            'Ace of Spades': 1, '2 of Spades': 2, '3 of Spades': 3, '4 of Spades': 4, '5 of Spades': 5,
            '6 of Spades': 6, '7 of Spades': 7, '8 of Spades': 8, '9 of Spades': 9, '10 of Spades': 10,
            'Jack of Spades': 10, 'Queen of Spades': 10, 'King of Spades': 10}
        
        self.player_hand=[]
        self.dealer_hand=[]
        self.total_amount=1000
        self.bet_amount=0
    
    def set_bet_amount(self,bet_amount):
        if bet_amount<=self.total_amount:
            self.bet_amount=bet_amount
            print(f"Total balance is {self.total_amount} and Bet amount is \033[1;33m{self.bet_amount}\033[0m\n")
        else:
            print("Insufficient balance\n")
        
    def draw_playercard(self):
        card=random.choice(list(self.deck.items()))
        del self.deck[card[0]]  # Remove the selected card from the deck
        return card
    
    def draw_dealercard(self):
        card=random.choice(list(self.deck.items()))
        del self.deck[card[0]]  # Remove the selected card from the deck
        return card
    
    def players_hand(self,bet_amount):
        
        self.player_hand=[self.draw_playercard(),self.draw_playercard()] 
        self.player_hand_value=[card[1] for card in self.player_hand]
        self.player_hand_key=[card[0] for card in self.player_hand]
        
        self.dealer_hand=[self.draw_dealercard()] 
        self.dealer_hand_value=[card[1] for card in self.dealer_hand]
        self.dealer_hand_key=[card[0] for card in self.dealer_hand]
       
        totalsum_player=sum(self.player_hand_value)                                 
        print(f"Player face up cards is \033[1;31m{self.player_hand_key}\033[0m\n")
        print(f"player face up card sum is {totalsum_player}\n")
        
        totalsum_dealer=sum(self.dealer_hand_value)
        print(f"Dealer face up cards is \033[1;31m{self.dealer_hand_key}\033[0m\n")
        print(f"Dealer face up card sum is {totalsum_dealer}\n")
    
        if sum(self.player_hand_value)==21:
            print("\033[1;35mBLACKJACK: Player wins\033[0m\n")
            self.total_amount+=(bet_amount*3)
            
            print(f"Total amount after this game: \033[1;32m{self.total_amount}\033[0m\n")
            self.play_again_prompt() 
            
        else:
            self.hit_stand()
            
    def play_again_prompt(self):
        
        while True:
            
            print("\33[1;34mDo you want to play again?\033[0m\n")
            reply=input("Enter your response with Y or N: \n").strip().lower()
            
            if reply=='y':
                bet_amount = int(input("Enter the bet amount: \n"))  # Ask for bet_amount again
                if bet_amount <= self.total_amount:   
                    player1 = BlackJack_game() # Create a new instance
                    player1.total_amount = self.total_amount
                    player1.set_bet_amount(bet_amount)  # Set the bet amount
                    player1.players_hand(bet_amount)  # Start a new game
                    return bet_amount
                else:
                    print("Insufficient balance\n")
                    continue
                
            elif reply=='n':
                print("\33[1;33mGAME OVER\033[0m")
                break
            
            else:
                print("\033[1;34mInvalid input. Please enter Y or N.\n\033[0m")
                return self.play_again_prompt()
            
    def hit_stand(self):
        while True:
            print("\033[1;34mDo you want to 'Hit' or 'Stand' ?\033[0m\n")
            reply=input("Enter either H or S: \n").strip().lower()

            if reply=='s':
            
                totalsum_player=sum(self.player_hand_value)
                
                print(f"player face up cards are \033[1;31m{self.player_hand_key}\033[0m\n")
                print(f"player face up card sum is {totalsum_player}\n")
                
                totalsum_dealer=sum(self.dealer_hand_value)
                
                while totalsum_dealer<=16:
                    new_draw_dealercard=self.draw_dealercard()
                    self.dealer_hand.append(new_draw_dealercard)
                    self.dealer_hand_key.append(new_draw_dealercard[0])
                    self.dealer_hand_value.append(new_draw_dealercard[1])
                    totalsum_dealer += new_draw_dealercard[1]
                    
                  
                print(f"Dealer face up cards are \033[1;31m{self.dealer_hand_key}\033[0m\n")
                print(f"Dealer face up card sum is {totalsum_dealer}\n") 
                 
                if totalsum_player>totalsum_dealer and totalsum_player<=21:
                    print("\033[1;35mPlayer wins\033[0m\n")
                    self.total_amount+=bet_amount
                    
                elif totalsum_dealer>21:
                    print("\033[1;35mDealer Busts && Player wins\033[0m\n")
                    self.total_amount+=bet_amount
                    
                elif totalsum_player==totalsum_dealer:
                    print("\033[1;35mPlayer Draw\033[0m\n")
                    
                else:
                    print("\033[1;35mPlayer Busts && Dealer wins\033[0m\n")
                    self.total_amount-=bet_amount
                    
                print(f"Total amount after this game: \033[1;32m{self.total_amount}\033[0m\n")    
                self.play_again_prompt()  
                
            elif reply=='h':
                
                new_draw_playercard=self.draw_playercard()
                self.player_hand.append(new_draw_playercard)
                self.player_hand_key.append(new_draw_playercard[0])
                self.player_hand_value.append(new_draw_playercard[1])  
                
                totalsum_player=sum(self.player_hand_value)
                
                print(f"player face up cards are \033[1;31m{self.player_hand_key}\033[0m\n")
                print(f"player face up card sum is {totalsum_player}\n")
                
                if totalsum_player>21:
                    print("\033[1;35mPlayer Busts\033[0m\n")
                    self.total_amount-=bet_amount
                    
                    print(f"Total amount after this game: \033[1;32m{self.total_amount}\033[0m\n")
                    self.play_again_prompt() 
                
                else:   
                    continue
            else:
                print("\033[1;34mInvalid reply.Please enter either 'H' for Hit or 'S' for Stand.\033[0m\n")
                continue
            break
          
    
print("\n\033[1mWelcome to BlackJack game!!!\033[0m\n")
player1=BlackJack_game()
print(f"Your total balance is \033[1;32m{player1.total_amount}\033[0m\n")
bet_amount=int(input("Enter the bet amount: \n"))
player1.set_bet_amount(bet_amount)
player1.draw_playercard()
player1.draw_dealercard()
player1.players_hand(bet_amount)
player1.play_again_prompt()
    
```
