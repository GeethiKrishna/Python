```python
import math
import random
from tkinter import *
from tkinter import Menu
from tkinter import ttk

# Global memory variable
memory = 0.0


# Function to update display with the button value
def update_display(value):
    display.insert(END, value)

# Function to evaluate the expression in the display
def evaluate_expression():
    try:
        result = str(eval(display.get()))
        display.delete(0, END)
        display.insert(END, result)
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")

# Function to clear the display
def clear_display():
    display.delete(0, END)

# Function to delete the last character in the display
def backspace():
    current_text = display.get()
    display.delete(0, END)
    display.insert(END, current_text[:-1])

def square_number():
    try:
        current_value = float(display.get()) if display.get() else 0.0  # Default to 0.0 if display is empty
        squared_value = current_value ** 2
        display.delete(0, END)
        display.insert(END, squared_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")


def calculate_square_root():
    try:
        current_value = float(display.get()) if display.get() else 0.0  # Default to 0.0 if display is empty
        sqrt_value = math.sqrt(current_value)
        display.delete(0, END)
        display.insert(END, sqrt_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")   
        
def calculate_reciprocal():
    try:
        current_value = float(display.get()) if display.get() else 0.0  # Default to 0.0 if display is empty
        reciprocal_value = 1 / current_value
        display.delete(0, END)
        display.insert(END, reciprocal_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except ZeroDivisionError:
        display.delete(0, END)
        display.insert(END, "Error: Division by zero")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")  
        
def calculate_percentage():
    try:
        current_value = float(display.get()) if display.get() else 0.0  # Default to 0.0 if display is empty
        percentage_value = current_value / 100
        display.delete(0, END)
        display.insert(END, percentage_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except ZeroDivisionError:
        display.delete(0, END)
        display.insert(END, "Error: Division by zero")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")       

def toggle_sign():
    try:
        current_value = float(display.get())
        toggled_value = -current_value
        display.delete(0, END)
        display.insert(END, toggled_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")
        
def clear_memory():
    global memory
    memory = 0.0

# Function to recall memory value to display
def recall_memory():
    global memory
    display.delete(0, END)
    display.insert(END, memory)

# Function to add current value to memory
def add_to_memory():
    try:
        global memory
        current_value = float(display.get())
        memory += current_value
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")

# Function to subtract current value from memory
def subtract_from_memory():
    try:
        global memory
        current_value = float(display.get())
        memory -= current_value
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")   
        
def store_in_memory():
    try:
        global memory
        current_value = float(display.get())
        memory = current_value
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")
        
def calculate_log():
    try:
        current_value = float(display.get()) if display.get() else 0.0  # Default to 0.0 if display is empty
        log_value = math.log(current_value, 10)  # Example with base 10
        display.delete(0, END)
        display.insert(END, log_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")
        
def calculate_ln():
    try:
        current_value = float(display.get()) if display.get() else 0.0  # Default to 0.0 if display is empty
        ln_value = math.log(current_value)
        display.delete(0, END)
        display.insert(END, ln_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")
      
        
def exponential():
    try:
        current_value = float(display.get()) if display.get() else 0.0  # Default to 0.0 if display is empty
        exp_value = math.exp(current_value)
        display.delete(0, END)
        display.insert(END, exp_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")    
        
def calculate_mod():
    try:
        # Split the input on the modulo operator
        values = display.get().split('%')
        
        # Convert the split values to floats and calculate the modulus
        if len(values) == 2:
            num1 = float(values[0])
            num2 = float(values[1])
            mod_value = num1 % num2
            display.delete(0, END)
            display.insert(END, mod_value)
        else:
            display.delete(0, END)
            display.insert(END, "Error")
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error") 
        
def absolute_value():
    try:
        current_value = float(display.get()) if display.get() else 0.0  # Default to 0.0 if display is empty
        abs_value = abs(current_value)
        display.delete(0, END)
        display.insert(END, abs_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error") 
        
def calculate_factorial():
    try:
        current_value = int(display.get())  # Get the current value from the display
        if current_value < 0:
            raise ValueError("Negative value")
        factorial_value = math.factorial(current_value)
        display.delete(0, END)
        display.insert(END, factorial_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error") 
        
def calculate_power_of_10():
    try:
        current_value = float(display.get()) if display.get() else 0.0  # Default to 0.0 if display is empty
        power_of_10_value = 10 ** current_value
        display.delete(0, END)
        display.insert(END, power_of_10_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")  
        
def calculate_power():
    try:
        base = float(display.get().split('^')[0])  # Assuming input format is "base^exponent"
        exponent = float(display.get().split('^')[1])
        power_value = base ** exponent
        display.delete(0, END)
        display.insert(END, power_value)
    except (ValueError, IndexError):
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")        
        
        
def insert_e():
    display.delete(0, END)
    display.insert(END, str(math.e)) 
    
def insert_pi():
    display.delete(0, END)
    display.insert(END, str(math.pi))
    
# Function to calculate sine
def calculate_sin():
    try:
        current_value = float(display.get())
        sin_value = math.sin(math.radians(current_value))  # Convert to radians if necessary
        display.delete(0, END)
        display.insert(END, sin_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")

# Function to calculate cosine
def calculate_cos():
    try:
        current_value = float(display.get())
        cos_value = math.cos(math.radians(current_value))  # Convert to radians if necessary
        display.delete(0, END)
        display.insert(END, cos_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")

# Function to calculate tangent
def calculate_tan():
    try:
        current_value = float(display.get())
        tan_value = math.tan(math.radians(current_value))  # Convert to radians if necessary
        display.delete(0, END)
        display.insert(END, tan_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")

# Function to calculate hyperbolic sine
def calculate_hyp():
    try:
        current_value = float(display.get())
        hyp_value = math.sinh(current_value)
        display.delete(0, END)
        display.insert(END, hyp_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")

# Function to calculate secant
def calculate_sec():
    try:
        current_value = float(display.get())
        sec_value = 1 / math.cos(math.radians(current_value))  # Convert to radians if necessary
        display.delete(0, END)
        display.insert(END, sec_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except ZeroDivisionError:
        display.delete(0, END)
        display.insert(END, "Error: Division by zero")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")

# Function to calculate cosecant
def calculate_csc():
    try:
        current_value = float(display.get())
        csc_value = 1 / math.sin(math.radians(current_value))  # Convert to radians if necessary
        display.delete(0, END)
        display.insert(END, csc_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except ZeroDivisionError:
        display.delete(0, END)
        display.insert(END, "Error: Division by zero")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")

# Function to calculate cotangent
def calculate_cot():
    try:
        current_value = float(display.get())
        cot_value = 1 / math.tan(math.radians(current_value))  # Convert to radians if necessary
        display.delete(0, END)
        display.insert(END, cot_value)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except ZeroDivisionError:
        display.delete(0, END)
        display.insert(END, "Error: Division by zero")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")   
        
# Function to handle the '2ᶮᵈ' operation
def calculate_second_power():
    try:
        current_value = float(display.get())
        result = current_value ** 2
        display.delete(0, END)
        display.insert(END, result)
    except ValueError:
        display.delete(0, END)
        display.insert(END, "Error")
    except Exception as e:
        display.delete(0, END)
        display.insert(END, "Error")   
        
# Absolute value function
def absolute(x):
    return abs(x)

# Floor function
def floor(x):
    return math.floor(x)

# Ceiling function
def ceil(x):
    return math.ceil(x)

# Random number function
def random_number():
    return round(random.random(), 6)  # Generating a random number between 0 and 1

# Convert radians to degrees, minutes, seconds
def rad_to_dms(x):
    degrees = math.degrees(x)
    deg = math.floor(degrees)
    minutes = (degrees - deg) * 60
    min = math.floor(minutes)
    seconds = (minutes - min) * 60
    return f"{deg}° {min}' {seconds:.2f}\""

# Convert degrees to radians
def deg_to_rad(x):
    return math.radians(x)
    
    
def select_option(option):
    print(f"Selected option: {option}")
    

window = Tk()
window.title("Calculator")
window.geometry("320x515")

# Create a custom style for the notebook tabs
style = ttk.Style()
style.configure('Custom.TNotebook.Tab', foreground='#0000ff', font=('Arial', 12)) 

# Create the Notebook widget
tab_control = ttk.Notebook(window, style='Custom.TNotebook')

# Create tabs
tab1 = ttk.Frame(tab_control)
tab2 = ttk.Frame(tab_control)

tab_control.add(tab1, text="Standard")
tab_control.add(tab2, text="Scientific")

tab_control.pack(expand=True, fill='both')

# Create the display area for calculator
display = Entry(window, font=("Calibri", 20), justify='right', bd=10, insertwidth=2, width=34, borderwidth=4)
display.place(x=2, y=40, width=310, height=90)

#create buttons for standard calculator
btnA=Button(tab1,text="MC",font=("calibri",12),width=5,height=1,bg="light blue",command=clear_memory)
btnA.place(x=3,y=120)
btnB=Button(tab1,text="MR",font=("calibri",12),width=5,height=1,bg="light blue",command=recall_memory)
btnB.place(x=55,y=120)
btnC=Button(tab1,text="M+",font=("calibri",12),width=5,height=1,bg="light blue",command=add_to_memory)
btnC.place(x=107,y=120)
btnD=Button(tab1,text="M-",font=("calibri",12),width=5,height=1,bg="light blue",command=subtract_from_memory)
btnD.place(x=159,y=120)
btnE=Button(tab1,text="MS",font=("calibri",12),width=5,height=1,bg="light blue")
btnE.place(x=211,y=120)
btnF=Button(tab1,text="M",font=("calibri",12),width=5,height=1,bg="light blue")
btnF.place(x=263,y=120)

buttons_standard_sym = [("%", 3, 160), ("CE", 82, 160), ("C", 162, 160), ("⌫", 242, 160),
    ("1/x", 3, 215), ("x²", 82, 215), ("²√x", 162, 215), ("/", 242, 215),("*", 242, 270),
     ("-", 242, 325),("+", 242, 380)]

buttons_standard_num = [("7", 3, 270), ("8", 82, 270), ("9", 162, 270),
    ("4", 3, 325), ("5", 82, 325), ("6", 162, 325),
    ("1", 3, 380), ("2", 82, 380), ("3", 162, 380),("+/-", 3, 435), ("0", 82, 435), (".", 162, 435)]
    
btn_st=Button(tab1,text="=",font=("calibri",12),width=8,height=2,bg="blue",fg="white",command=evaluate_expression)
btn_st.place(x=242,y=435)

for text, x, y in buttons_standard_sym + buttons_standard_num:
    
    if text == "C" or text == "CE":
        command = clear_display
    elif text == "⌫":
        command = backspace
    elif text == "+/-":
        command = toggle_sign    
    elif text == "%":
        command = calculate_percentage  
    elif text == "1/x":
        command = calculate_reciprocal    
    elif text == "²√x":
        command = calculate_square_root
    elif text == "x²":
        command = square_number
    else:
        command = lambda t=text: update_display(t)

    bg_color = "light grey" if (text, x, y) in buttons_standard_sym else "white"
    
    Button(tab1, text=text, font=("Calibri", 12), width=8, height=2, bg=bg_color, command=command).place(x=x, y=y)


# Create buttons for Scientific calculator tab

btn_scA=Button(tab2,text="MC",font=("calibri",12),width=5,height=1,bg="light blue",command=clear_memory)
btn_scA.place(x=3,y=113)
btn_scB=Button(tab2,text="MR",font=("calibri",12),width=5,height=1,bg="light blue",command=recall_memory)
btn_scB.place(x=55,y=113)
btn_scC=Button(tab2,text="M+",font=("calibri",12),width=5,height=1,bg="light blue",command=add_to_memory)
btn_scC.place(x=107,y=113)
btn_scD=Button(tab2,text="M-",font=("calibri",12),width=5,height=1,bg="light blue",command=subtract_from_memory)
btn_scD.place(x=159,y=113)
btn_scE=Button(tab2,text="MS",font=("calibri",12),width=5,height=1,bg="light blue")
btn_scE.place(x=211,y=113)
btn_scF=Button(tab2,text="M",font=("calibri",12),width=5,height=1,bg="light blue")
btn_scF.place(x=263,y=113)

buttons_scientific_sym = [("2ᶮᵈ",2,150),("π",67,150),("e",125,150),("CE",185,150),("⌫",248,150),\
                      ("x²",2,200),("1/x",67,200),("|x|",125,200),("exp",185,200),("mod",248,200),\
                      ("²√x",2,250),("(",67,250),(")",125,250),("n!",185,250),("/",248,250),\
                      ("x^y",2,300),("*",248,300),\
                      ("10^x",2,350),("-",248,350),\
                      ("log",2,400),("+",248,400),\
                     ("ln",2,450)]

btn_sc=Button(tab2,text="=",font=("calibri",12),width=7,height=2,bg="blue",command=evaluate_expression)
btn_sc.place(x=248,y=450)

buttons_scientific_num = [("7",67,300),("8",125,300),("9",182,300),("4",67,350),("5",125,350),("6",182,350),\
                      ("1",67,400),("2",125,400),("3",182,400),\
                     ("+/-",67,450),("0",125,450),(".",182,450)]

for text, x, y in buttons_scientific_sym + buttons_scientific_num:
    
    if text == "C" or text == "CE":
        command = clear_display
    elif text == "⌫":
        command = backspace
    elif text == "+/-":
        command = toggle_sign    
    elif text == "%":
        command = calculate_percentage    
    elif text == "1/x":
        command = calculate_reciprocal    
    elif text == "²√x":
        command = calculate_square_root
    elif text == "x²":
        command = square_number 
    elif text == "log":
        command = calculate_log   
    elif text == "ln":
        command = calculate_ln     
    elif text == "exp":
        command = exponential 
    elif text == "mod":
        command = calculate_mod 
    elif text == "|x|":
        command = absolute_value 
    elif text == "n!":
        command = calculate_factorial 
    elif text == "10^x":
        command = calculate_power_of_10
    elif text == "x^y":
        command = calculate_power    
    elif text == "e":
        command = insert_e    
    elif text == "π":
        command = insert_pi    
    else:
        command = lambda t=text: update_display(t)

    bg_color = "light grey" if (text, x, y) in buttons_scientific_sym else "white"
    
    Button(tab2, text=text, font=("Calibri", 12), width=7, height=2, bg=bg_color, command=command).place(x=x, y=y)
    
    
# Create the main menu
main_menu = Menu(tab2)

# Create the Trigonometry submenu
trigonometry_menu = Menu(main_menu, tearoff=0,font=("Calibri", 12))
trigonometry_menu.add_command(label="2ᶮᵈ",command=calculate_second_power)
trigonometry_menu.add_command(label="sin",command=calculate_sin)
trigonometry_menu.add_command(label="cos",command=calculate_cos)
trigonometry_menu.add_command(label="tan",command=calculate_tan)
trigonometry_menu.add_command(label="hyp",command=calculate_hyp)
trigonometry_menu.add_command(label="sec",command=calculate_sec)
trigonometry_menu.add_command(label="csc",command=calculate_csc)
trigonometry_menu.add_command(label="cot",command=calculate_cot)
main_menu.add_cascade(label="⊿ Trigonometry ▼", menu=trigonometry_menu)

# Create the Function submenu
function_menu = Menu(main_menu, tearoff=0, font=("Calibri", 12))
function_menu.add_command(label="|x|",command=absolute(x))
function_menu.add_command(label="⌊x⌋",command=floor(x))
function_menu.add_command(label="⌈x⌉",command=ceil(x))
function_menu.add_command(label="rand",command=random_number())
function_menu.add_command(label="→dms",command=rad_to_dms(x))
function_menu.add_command(label="→deg",command=deg_to_rad(x))
main_menu.add_cascade(label="Ƒ Function ▼", menu=function_menu)

# Function to update the menu based on the selected tab
def on_tab_change(event):
    selected_tab = event.widget.index("current")
    if selected_tab == 1:  # Scientific tab
        window.config(menu=main_menu)
    else:  # Standard tab
        window.config(menu='')
               

# Bind the tab change event
tab_control.bind("<<NotebookTabChanged>>", on_tab_change)



window.config(menu=main_menu)
window.mainloop()
```
